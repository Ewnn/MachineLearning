```python
#data generation
import numpy as np
X=np.array([0,3,6,8]) #coordonnées du point en X
Y=np.array([35,45,65,80]) #coordonnées du point en Y

#graphical window
import matplotlib.pyplot as plt 
plt.plot(X,Y,'*') #point of form
plt.xlabel("Years of exeprience |  our explicative variable 'x' ")
plt.ylabel("Salary | Target variable 'y'")
plt.title("Scratter plot")
plt.savefig("./intuitive_scatter.png")

#legend of graphics
def reg_plot(x,y,m):
    plt.scatter(x,y,c='blue',label="our data")
    plt.plot(x, m.predict(x.reshape(-1, 1)), color='red',label="prediction curve")
    plt.xlabel("explicative variable 'x' ")
    plt.ylabel("target variable 'y'")
    plt.legend()
    return None 

#curve draw
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(X.reshape(-1, 1),Y) #.fit => trace la droite qui fait la somme des ecarts au carré
reg_plot(X,Y,linear_model)
plt.savefig("./approche_intuitive.png")
```

```python
import pandas as pd 
df = pd.read_csv("data/Salary_Data.csv", sep=',')
df.head()
```

```python
df=df[["YearsExperience","Salary"]]
X=df.YearsExperience
Y=df.Salary
plt.plot(X,Y,'*')
plt.xlabel("Years of exeprience |  our explicative variable 'x' ")
plt.ylabel("Salary | Target variable 'y'")
plt.title("Years of experience vs Salary")
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(np.array(X).reshape(-1, 1),np.array(Y))
reg_plot(np.array(X),np.array(Y),linear_model)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
Y_pred = linear_model.predict(np.array(X).reshape(-1, 1))
print("Mean squared error: %.2f" 
      % mean_squared_error(Y, Y_pred)) # somme de tout les points (Ytrue - Ypred)² = le résultat est en € au carré
print("Root mean squared error: %.2f" 
      % np.sqrt(mean_squared_error(Y, Y_pred))) # racine de Mean squared error, qui nous donne la marge d'erreur en €
print("R square: %.2f"% r2_score(Y, Y_pred)) # coralité
```
