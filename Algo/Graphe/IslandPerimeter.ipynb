```python
class Solution(object):
    def islandPerimeter(self, grid):
        rows = len(grid)
        cols = len(grid[0])
        perimeter = 0

        for row in range(rows):
            for col in range(cols):
                if grid[row][col] == 1:
                    # Ajoute 4 pour chaque cellule de terre
                    perimeter += 4
                    
                    # Soustrait 2 pour chaque voisin de terre à droite
                    if col < cols - 1 and grid[row][col + 1] == 1:
                        perimeter -= 2
                        
                    # Soustrait 2 pour chaque voisin de terre en dessous
                    if row < rows - 1 and grid[row + 1][col] == 1:
                        perimeter -= 2

        return perimeter
```
