```python
def mergeTwoLists(list1, list2):
        fakenode = ListNode()
        stock = fakenode

        while list1 and list2:
            if list1.val < list2.val:
                stock.next = list1
                list1 = list1.next
            else:
                stock.next = list2
                list2 = list2.next
            stock = stock.next
        
        if list1:
            stock.next = list1
        elif list2:
            stock.next = list2
        
        return fakenode.next
```
