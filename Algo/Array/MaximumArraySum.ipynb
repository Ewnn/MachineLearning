```python
#Version O(n²)
def MaxSum():
    t = [2, 3, -8, 7, -1, 2, 3]
    res = float('-inf')
    for i in range(len(t)):
        current_sum = sum(t[i:]) 
        if current_sum > res:
            res = current_sum  
    return res

MaxSum()
```

```python
#Version O(n)
def maxSum():
    t = [2, 3, -8, 7, -1, 2, 3]
    res = float('-inf')
    current_sum = 0
    maxS = []
    temp = []

    for num in t:
        if current_sum + num > num:
            current_sum += num
            temp.append(num)
        else:
            current_sum = num
            temp = [num]

        if current_sum > res:
            res = current_sum
            maxS = temp[:]

    return res, maxS

print(maxSum())
```
