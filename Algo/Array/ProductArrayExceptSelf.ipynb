```python
def productExceptSelf(nums):
    n = len(nums)
    answer = [1] * n
    
    gauche = 1
    for i in range(n):
        answer[i] = gauche
        gauche *= nums[i]
    
    droite = 1
    for i in range(n - 1, -1, -1):
        answer[i] *= droite
        droite *= nums[i]
    
    return answer

productExceptSelf(nums=[1,2,3,4])
```
