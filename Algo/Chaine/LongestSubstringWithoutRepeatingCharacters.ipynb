```python
def lengthOfLongestSubstring(s):
    start = 0
    max_len = 0
    used_letter = set()

    for letter in range(len(s)):
        while s[letter] in used_letter:
            used_letter.remove(s[start])
            start += 1
        used_letter.add(s[letter])
        max_len = max(max_len, letter - start + 1)

    return max_len

lengthOfLongestSubstring(s="abcabbcacdeabcdef")
```
