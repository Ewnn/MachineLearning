```python
def subarraySum(nums, k):
    count = 0
    cumulative_sum = 0
    sum_map = {0: 1}
    
    for num in nums:
        cumulative_sum += num
        
        if (cumulative_sum - k) in sum_map:
            count += sum_map[cumulative_sum - k]

        if cumulative_sum in sum_map:
            sum_map[cumulative_sum] += 1
        else:
            sum_map[cumulative_sum] = 1
    
    return count    

subarraySum(nums=[1,1,1], k=2)
```
